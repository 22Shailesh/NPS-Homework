- Define random walk in 2D.
- Create a random walk data and save the data in a file (json/csv).
- Create a plot for random walk in 2D as save the plot as (pdf/png) file. 

<u>*Definition of random walk in 2D*</u>:  Random walk in 2D refers to the motion of a subject that can move one step in right, left, forward, and backward direction with equal probability. Usually, the lengh of each step is assumed to be constant.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = 0
y = 0


X = [x]
Y = [y]

t = 1

N = 20000



while t < N:
    direction = random.choice(['parallel','perpendicular'])
    if direction == 'parallel':
        r = random.choice([-1,1])
        x = X[t-1]+r
        y = Y[t-1]
        X.append(x)
        Y.append(y)
    if direction == 'perpendicular':
        r = random.choice([-1,1])
        y = Y[t-1]+r
        x = X[t-1]
        X.append(x)
        Y.append(y)
        
    
    t = t+1
```

```python
data = {"x":X,"y":Y}
import json as json
with open ('data/random_walk_2D.json','w') as f:
    json.dump(data,f)
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df.to_csv('data/random_walk_2D_DF.csv')
```

```python
plt.figure(figsize = [8,8])
plt.plot(X,Y,color = 'blue')
plt.savefig('plot/random_walk_2D.pdf')
plt.savefig('plot/random_walk_2D.png')
```
