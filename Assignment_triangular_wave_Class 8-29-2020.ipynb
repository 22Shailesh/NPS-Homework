Develop the Fourier series of Triangular wave following the above method:

$$f(x) = \dfrac{8}{\pi^2} \sum\limits_{n = 1, 3, 5, ....}^{\inf}{\dfrac{(-1)^{(n-1)/2}}{n^2} \sin{\bigg(\dfrac{n\pi x}{L}\bigg)}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

**Individual basis**

```python
L = 1
x_step = 1000
x = np.arange(0, 2, 10/x_step)

plt.figure(figsize = [15,4])

for n in [i for i in range(10) if i%2 != 0]:
    fi = (8/np.pi**2)*(-1)**(0.5*(n - 1))/n**2*np.sin(n*np.pi*x/L)
    plt.plot(x, fi)
```

#### Subsequent development

```python
L = 1
x_step = 1000
x = np.arange(0, 2, 10/x_step)

plt.figure(figsize = [15,4])

for n in [i for i in range(10) if i%2 != 0]:
    if n == 1:
        fi = (8/np.pi**2)*(-1)**(0.5*(n - 1))/n**2*np.sin(n*np.pi*x/L)
        plt.plot(x, fi, label = 'n = ' + str(n))
        ff = fi
    else:
        fi = ff + (8/np.pi**2)*(-1)**(0.5*(n - 1))/n**2*np.sin(n*np.pi*x/L)
        plt.plot(x, fi, label = 'n = ' + str(n))
        ff = fi
        
plt.legend()
plt.xlabel('x')
plt.ylabel('fx')
plt.show()
```

#### Sum of all n-terms

```python
L = 1
x_step = 1000
x = np.arange(0, 10, 10/x_step)

plt.figure(figsize = [15,4])

fx = sum([(8/np.pi**2)*(-1)**(0.5*(n - 1))/n**2*np.sin(n*np.pi*x/L) for n in [i for i in range(200) if i%2 != 0]])
plt.plot(x, fx, label = 'triangular wave')
        
plt.legend()
plt.xlabel('x')
plt.ylabel('fx')
plt.show()
```
