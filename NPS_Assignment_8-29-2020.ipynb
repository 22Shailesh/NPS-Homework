1. Write a python function creating a random oscillation with amplitude in range (1, 5) and frequency in range (0.25, 1)
2. Create 20 different oscillations by implementing python function defined above and implement superposition principle to obtain single resultant solution. 

### Solution of 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def random_oscillation():
    return [a*np.sin(2*np.pi*f*t) for a, f in zip(amplitude, frequency)]
```

```python
t = np.arange(0, 15, 0.01) # define time duration with a finite step size

amplitude = [i for i in np.arange (1, 5, (5 - 1)/4)] # four different amplitudes
frequency = [i for i in np.arange (0.25, 1, (1 - 0.25)/4)] # four different frequencies

y = random_oscillation() # implement the function to obtain a list of wave displacement values.
plt.figure(figsize = [15, 8])
plt.plot(t, y[0], 'k', lw = 3, label = 'amplitude = '+str(amplitude[0])+' cm, '+'frequency = '+str(frequency [0])+' Hz')
plt.plot(t, y[1], 'b', lw = 3, label = 'amplitude = '+str(amplitude[1])+' cm, '+'frequency = '+str(frequency [1])+' Hz')
plt.plot(t, y[2], 'g', lw = 3, label = 'amplitude = '+str(amplitude[2])+' cm, '+'frequency = '+str(frequency [2])+' Hz')
plt.plot(t, y[3], 'r', lw = 3, label = 'amplitude = '+str(amplitude[3])+' cm, '+'frequency = '+str(frequency [3])+' Hz')
plt.legend(fontsize = 14, loc = 'upper right')
plt.xlabel('time (s)', fontsize = 20)
plt.ylabel('displacement (cm)', fontsize = 20)
plt.xticks(size = 18)
plt.yticks(size = 18)
plt.show()
```

### Solution of 2.

```python
t = np.arange(0, 15, 0.01) # define time duration with a finite step size
amplitude = [i for i in np.arange (1, 5, (5 - 1)/20)] # twenty different amplitudes
frequency = [i for i in np.arange (0.25, 1, (1 - 0.25)/20)] # twenty different frequencies
y = random_oscillation() # implement the function to obtain a list of wave displacement values.
y1 = sum(y)
plt.figure(figsize = [15, 8])
plt.plot(t, y1, color = 'maroon' , label = 'superposition of 20 different sine waves, each of different frequencies as well as amplitudes')
plt.legend(fontsize = 16)
plt.xlabel('time (s)', fontsize = 20)
plt.ylabel('displacement (cm)', fontsize = 20)
plt.xticks(size = 18)
plt.yticks(size = 18)
plt.show()
```
