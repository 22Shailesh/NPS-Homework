 1. Generate a random array of size 10 by 10 and convert it into pandas dataframe with column name and row name of your interest.
 2. Perform feature scaling (minimax scaler and standard scaler) by direct calculation of minimum, maximum, mean, and standard deviation from dataframe, array.
 3. Implement Scikit-learn preprocessing pipeline to perform feature scaling.

```python
import numpy as np
import pandas as pd
P = (100*np.random.rand(10, 10)).astype(int)
P
```

```python
df = pd.DataFrame(P, columns = ['black', 'brown', 'red', 'orange', 'yellow', 'green', 'blue', 'violet', 'grey', 'white'],\
                 index = ['Bic', 'Camlin', 'Cross', 'Dunhill', 'Hero', 'Hunt', 'Pilot', 'Reynolds', 'Waterman', 'Zebra'])
df
```

**Feature Scaling using mini-max scaling**

```python
P_mm = (P - P.min())/(P.max()-P.min())
P_mm
```

```python
from sklearn import preprocessing
P_scaled = preprocessing.MinMaxScaler().fit(P).transform(P)
P_scaled
```

**Feature Scaling using standard scaling**

```python
P_ss = (P-P.mean())/P.std()
P_ss
```

```python
P_scaled = preprocessing.StandardScaler().fit(P).transform(P)
P_scaled
```
